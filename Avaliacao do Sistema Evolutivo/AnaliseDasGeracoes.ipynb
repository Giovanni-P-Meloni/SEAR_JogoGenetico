# Análise das gerações

O primeiro teste contou com 100 gerações. Cada população possui 10 individuos que atiram 3 vezes cada.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

results = pd.read_csv('df_v2.0.csv')
meanFit = pd.read_csv('meanFit.csv')
```

analisando o formato dos dados

```python
result = pd.merge(results, meanFit, on='Generation')
result.info()
```

```python
#plt.figure()
fig, axs = plt.subplots(figsize=(8,4))
result.plot(ax=axs, x='Generation', y='Fitness')
result.plot(ax=axs, x='Generation', y='Mean Fitness', secondary_y=True)
plt.show()
#plt.ylim(0, 550)
```

```python
fig, axs = plt.subplots(figsize=(8,4))
result.plot(ax=axs, x='Generation', y='Fitness')
#result.plot(ax=axs, x='Generation', y='Mean Fitness', secondary_y=True)
result.plot(ax=axs, x='Generation', y='Mutation Rate', secondary_y=True)
plt.show()
```
